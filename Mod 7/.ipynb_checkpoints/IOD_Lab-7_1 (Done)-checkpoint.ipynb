## Now to maybe standardize the scale of the variables:

```python
#Define out variables and target:

X = df.drop('whether he/she donated blood in March 2007',
            axis = 1) #Remember positional arguments can come before keyword arguments, but not other way round

y = df['whether he/she donated blood in March 2007']

```

```python
#Scale the features:

from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()

scaled_X = pd.DataFrame(Scaler.fit_transform(X),
                        columns = X.columns)
```

```python
print('Shape of features:',scaled_X.shape)
print('Shape of target:',y.shape)
```

# Step 7: Split the Data

Need to check for **Supervised** modelling:
- Number of known cases or observations
- Define the split in Training/Test or Training/Validation/Test and their proportions

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.25, random_state=42)

```

```python
y_train.astype(int)
```

# Step 8: Define a Model

Define the model and its hyper-parameters.

Consider the parameters and hyper-parameters of each model at each (re)run and after checking the efficiency of a model against the training and test datasets.

```python
#Try support vector machine:

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

#Use GridSearchCV to find optimal parameters:

SVC_cv = GridSearchCV(estimator = SVC(),
                      param_grid = {'C': [0.1,0.5,1,5,10],
                                    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
                                   }
                     )
SVC_cv.fit(X_train,y_train.astype('int'))

#Print optimal parameters found:
print('Best parameters for Support Vector Classifier:',SVC_cv.best_params_)

#Predict values:
SVC_predictions = SVC_cv.predict(X_test)
```

```python
#Try K-Nearest Neighbors:

from sklearn.neighbors import KNeighborsClassifier

KNN_cv = GridSearchCV(estimator = KNeighborsClassifier(),
                      param_grid = {'n_neighbors':range(1,40)
                                   } 
                     )
#Fit model
KNN_cv.fit(X_train,y_train.astype(int))

#Print optimal parameters:
print('Optimal parameters for K-Nearest Neighbors Classifier:', KNN_cv.best_params_)

#Predict values
KNN_predictions = KNN_cv.predict(X_test)
```

```python
#Try Logistic Regression:

from sklearn.linear_model import LogisticRegression

Logmodel_cv = GridSearchCV(estimator = LogisticRegression(),
                           param_grid = {'C': [0.1,0.5,1,5,10],
                                         'solver': ['newton-cg','sag','lbfgs']
                                        }
                          )
#Fit model:
Logmodel_cv.fit(X_train,y_train.astype(int))

#Print best parameters:
print(Logmodel_cv.best_params_)

#Predict values:
Logmodel_predictions = Logmodel_cv.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

Tree = DecisionTreeClassifier()

Tree.fit(X_train,y_train.astype(int))

Tree_predictions = Tree.predict(X_test)

print('Decision tree accuracy:',accuracy_score(y_test.astype(int),Tree_predictions))
print('Accuracy score for Support Vector Machines Classifier:', accuracy_score(y_test.astype(int), SVC_predictions))
print('Accuracy score for K-Nearest Neighbors Classifier:', accuracy_score(y_test.astype(int), KNN_predictions))
print('Accuracy score for Logistic Regression:', accuracy_score(y_test.astype(int), Logmodel_predictions))
```
