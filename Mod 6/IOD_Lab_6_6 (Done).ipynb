##### 4.1 Univariate Selection

Univariate feature selection works by selecting the best features based on univariate statistical tests. It can be seen as a preprocessing step to an estimator. Scikit-learn exposes feature selection routines as objects that implement the transform method:

- SelectKBest removes all but the  highest scoring features
- Use sklearn.feature_selection.chi2 as score function
    > Recall that the chi-square test measures dependence between stochastic variables, so using this function “weeds out” the features that are the most likely to be independent of class and therefore irrelevant for classification.

##### 4.2 Recursive feature elimination

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coef_ attribute or through a feature_importances_ attribute. Then, the least important features are pruned from current set of features.That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

More Reads:
[Recursive feature elimination](https://scikit-learn.org/stable/modules/feature_selection.html)

- Use RFE to extract feature
    - use LogisticRegression as estimator
    - Number of n_features_to_select as of your choice
- Fit X, y to RFE
- Find Selected Features

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#Estimator is Log regression, to select 4 features

rfe = RFE(estimator = LogisticRegression(), 
          n_features_to_select = 4) 
```

```python
#Split data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Try Logistic Regression with GridSearchCV

from sklearn.linear_model import LogisticRegression

logmodel = GridSearchCV(estimator = LogisticRegression(),
                        param_grid = {'C':range(1,5)}
                       )

logmodel.fit(X_train,y_train)

print('Best parameter values for log model:',logmodel.best_params_)

log_predictions = logmodel.predict(X_test)

```

```python
#Try K-Nearest Neighbors with GridSearchCV 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

#Create GridSearch model using KNN classifier to find optimal k-value
KNNmodel_cv = GridSearchCV(estimator = KNeighborsClassifier(),
                           param_grid = {'n_neighbors' : range(1,20)} )

#Fit GridSearch model 
KNNmodel_cv.fit(X_train,y_train)

print('Best k-value found by Cross-Validation:',KNNmodel_cv.best_params_)

KNN_predictions = KNNmodel_cv.predict(X_test)
```

```python
#Try Support Vector Machines with GridSearchCV

from sklearn.svm import SVC

SVM_model_cv = GridSearchCV(estimator = SVC(),
                            param_grid = {'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
                                          'C': range(1,3)})
SVM_model_cv.fit(X_train,y_train)

print('Best parameters for Support Vector Classifier:', SVM_model_cv.best_params_)

SVM_predictions = SVM_model_cv.predict(X_test)
```

```python
#Evaluate classifier metrics:

from sklearn.metrics import accuracy_score

print('Accuracy score for Logistic Regression:', accuracy_score(y_test,log_predictions))
print('Accuracy score for K-Nearest Neighbors:', accuracy_score(y_test,KNN_predictions))
print('Accuracy score for Support Vector Machine:', accuracy_score(y_test, SVM_predictions))
print('Note: These classifiers use all features, not only K-Best')
```
