### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.
2. Put the features on the same scale.

```python
X = df.drop(['Sex','Sport','Unnamed: 0'],axis = 1)
```

```python
from sklearn import preprocessing

x = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
scaledvalues = min_max_scaler.fit_transform(x)
X_scaled = pd.DataFrame(scaledvalues, columns = X.columns )

X_scaled.head()
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Lets try using 2, 4. Run the clustering algorithm with each of those k values.

```python
from sklearn.cluster import KMeans

model_2 = KMeans(2).fit(X_scaled)
model_4 = KMeans(4).fit(X_scaled)

```

### 5. Get the labels for the models

```python
labels_k_2 = model_2.labels_
labels_k_4 = model_4.labels_
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero. 

```python
inertias = []

for i in range(1,20):

    model = KMeans(i).fit(X_scaled)
    inertias.append(model.inertia_)



```

Below is a plot of sum of squared distances for k in the range specified above. If the plot looks like an arm, then the elbow on the arm is optimal k.

```python
plt.figure(figsize = (10,5))

plt.plot(inertias)


plt.ylabel('Inertia (Sum of squares)')
plt.xlabel('K-value')
plt.title('Elbow plot for choosing the right value for K')

plt.show()
```

### 7. Plot `pcBfat` vs. `wt` using the clusters from each model.

- Plot actual classification of pcBfat vs wt using hue sex
- Plot predicted classification of pcBfat vs wt using hue label

```python
#Adding label columns to dataframe

X_scaled['Labels (k=2)'] = labels_k_2
X_scaled['Labels (k=4)'] = labels_k_4

#Adding back sex and sport columns

X_scaled['Sex'] = df['Sex']
X_scaled['Sport'] = df['Sport']
```
